```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
results = pd.read_csv("../evaluation/results_german.csv", index_col=0)
results.index = results.index.astype(str)

print(results)

# Only plot if classes exist
classes = ["1", "2"]
existing = [cls for cls in classes if cls in results.index]

if existing:
    results.loc[existing, ["precision", "recall", "f1-score"]].plot(
        kind="bar",
        figsize=(6, 4),
        title="German Credit Risk - Metrics"
    )
    plt.ylabel("Score")
    plt.ylim(0, 1)
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No matching classes found in results.")
```
